```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# 1. Carregar os Dados
# X são as características (medidas), y é o rótulo (a espécie da flor)
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# 2. Dividir os Dados
# 70% para treino, 30% para teste.
# random_state=42 garante que a divisão seja sempre a mesma, 
# permitindo resultados reproduzíveis.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 6. Avaliar o Desempenho
# Comparamos as previsões (y_pred) com os rótulos reais (y_test)
acuracia = accuracy_score(y_test, y_pred)
print(f"--- Mini Projeto: Classificador de Flores Iris ---")
print(f"Acurácia do modelo KNN: {acuracia * 100:.2f}%")
```
